# Working of Keras

## Some of the key 'Technical Terms'

<table style="border: 1px solid black">
    <tr style="border: 1px solid black">
        <th style="float: left">Topic</th>
        <th style="border: 1px solid black; text-align: center">Details</th>
    </tr>
    <tr style="border: 1px solid black">
        <td style="float: left"> 
            <b>Feature: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            The input(s) to our model
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Examples: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            An input/output pair used for training
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Labels: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            The output of the model
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Layer: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            A collection of nodes connected together within a neural network.
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Model: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            The representation of your neural network
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Dense and Fully Connected (FC): </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            Each node in one layer is connected to each node in the previous layer.
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Weights and biases: </b>
        </td>
        <td style="border: 1px solid black; text-align: left"> 
            The internal variables of model
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Loss: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         The discrepancy between the desired output and the actual output
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>MSE: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         Mean squared error, a type of loss function that counts a small number of large discrepancies as worse than a large number of small ones.
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Gradient Descent: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         An algorithm that changes the internal variables a bit at a time to gradually reduce the loss function.
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Optimizer: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         A specific implementation of the gradient descent algorithm. (There are many algorithms for this. In this course we will only use the “Adam” Optimizer, which stands for ADAptive with Momentum. It is considered the best-practice optimizer.)
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Learning rate: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         The “step size” for loss improvement during gradient descent.
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Batch: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         The set of examples used during training of the neural network
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Epoch: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         A full pass over the entire training dataset
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Forward pass: </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         The computation of output values from input
        </td>
    </tr>
    <tr style="border: 1px solid black">
        <td style="border: 1px solid black; text-align: left"> 
            <b>Backward pass (backpropagation): </b>
        </td>
        <td style="border: 1px solid black; text-align: left">
         The calculation of internal variable adjustments according to the optimizer algorithm, starting from the output layer and working back through each layer to the input.
        </td>
    </tr>
</table>
